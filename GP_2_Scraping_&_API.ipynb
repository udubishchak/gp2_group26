## Оценивание и штрафы



Групповой проект оценивается исходя из устной защиты (презентации) командой выполненного ими проекта перед комиссией. Комиссия составляется из преподавателей курса, ассистентов курса, а также приглашенных экспертов из индустрии.

В рамках каждой защиты комиссия выставляет два типа оценок.

Первая оценка — это оценка для всего проекта в целом. Она ставится комиссией, исходя из глубины проработки командой предметной области, сложности применяемых в проекте инструментов, а также прочих аспектов, связанных с содержанием выполняемого в проекте ТЗ. При выставлении этой оценки комиссия ориентируется на критерии, более подробно представленные ниже в разделе "Критерии оценки проекта".

Вторая оценка — это индивидуальная оценка каждого участника команды. Именно эта оценка выставляется впоследствии в ведомость. Индивидуальная оценка участника команды определяется комиссией, исходя из того, насколько участник ориентируется в выполненном командой проекте, как он отвечает на адресованные персонально ему вопросы комиссии, насколько в принципе прослеживается его роль в проекте, а также исходя из других аспектов, связанных с участием конкретного человека в деятельности команды. Индивидуальная оценка не может превосходить оценку команды за проект в целом (первую оценку), однако, в случае когда участник спокойно и корректно отвечает на все адресованные ему вопросы, свободно ориентируется в материале и содержании выполненного проекта, а также когда его роль в работе команды не вызывает сомнений, — в таком случае, индивидуальная оценка участника становится равной оценке для всего проекта (т.е. в идеальном случае — вторая оценка равна первой оценке). Кроме того, в исключительных случаях, при демонстрации участником на защите превосходящих ожидания уровней понимания, участия и вовлечения в рамках проекта, комиссия может своим решением повысить индивидуальную оценку для этого участника на 0,5 или 1 балл относительно общей оценки за проект (первой оценки).

Во избежание субъективности и/или предвзятости оценки тех или иных членов комиссии в отношении тех или иных команд, а также каких-либо иных проявлений неравных условий — распределение команд между комиссиями осуществляется абсолютно рандомно с использованием средств библиотеки Numpy :)) — в качестве рэндом сида выступает дата, в которую производится распределение, например, `np.random.seed(21022025)`.

Обратите, пожалуйста, внимание на то, что бОльшая часть комиссий осуществляет приём проектов в очной форме, однако могут присутствовать и 1-2 комиссии, которые будут работать дистанционно с использованием средств видеосвязи. ## Извлечение данных путём работы с API

Ниже написаны коды для 6 запросов для сбора данных с сайта HH при помощи API. Каждый запрос выдает 2000 вакансий по городу Москва, но по 6 направлениям: разработка, аналитика, менеджмент, маркетинг, HR, PR.

добавим признак с ключевыми навыками. Последний признак занимает примерно треть кода, так как: В результате поиска (эндпоинт .../vacancies?text=...) ключевых навыков нет, а есть лишь упрощённый «snippet» (короткий фрагмент описания).
Ключевые навыки (key_skills) находятся только в детальной информации о вакансии, доступной по эндпоинту https://api.hh.ru/vacancies/{vacancy_id}.
Поэтому, чтобы собрать навыки, приходится для каждого job['id'] делать отдельный запрос.
Именно это и удлиняет код: дополнительный цикл с запросом, где мы формируем строку из ключевых навыков и добавляем её в итоговый DataFrame.

```python
# URL API HH.ru для получения вакансий
api_url = "https://api.hh.ru/vacancies"

# Функция для получения данных с нескольких страниц API, с учетом ограничения в 2000 записей
def get_multiple_pages_of_data(keyword, total_jobs=2000):
    all_jobs = []
    max_retrievable_jobs = 2000  # Ограничение API HH.ru
    page = 0

    while len(all_jobs) < min(total_jobs, max_retrievable_jobs):
        params = {
            'text': keyword,      # Параметр поиска
            'area': 1,            # Москва
            'per_page': 100,      # Количество вакансий на странице
            'page': page,         # Номер страницы
        }

        response = requests.get(api_url, params=params)

        if response.status_code == 200:
            data = response.json()
            all_jobs.extend(data['items'])
            print(f"Загружено {len(all_jobs)} вакансий (страница {page + 1})")
            page += 1

            if not data['items']:  # Проверка на пустую страницу (конец результатов)
                break

    print(f"Общее количество собранных вакансий: {len(all_jobs)}")
    return all_jobs


def convert_to_dataframe(jobs):
    job_data = []
    for job in jobs:
        # --- Дополнительный запрос для получения key_skills ---
        key_skills_str = 'Не указано'
        if 'id' in job:
            vacancy_id = job['id']
            detail_url = f"https://api.hh.ru/vacancies/{vacancy_id}"
            detail_response = requests.get(detail_url)

            if detail_response.status_code == 200:
                detail_data = detail_response.json()
                if 'key_skills' in detail_data and detail_data['key_skills']:
                    # Собираем навыки в одну строку, разделённую запятыми
                    key_skills_str = ", ".join(skill['name'] for skill in detail_data['key_skills'])

        # Формируем словарь с прежними полями + новый столбец 'Key_skills'
        job_info = {
            'Vacancy': job['name'],
            'Salary_from': job['salary']['from'] if job['salary'] else 'Не указана',
            'Salary_to': job['salary']['to'] if job['salary'] else 'Не указана',
            'Salary_currency': job['salary']['currency'] if job['salary'] else 'Не указана',
            'Employer_name': job['employer']['name'] if job['employer'] else 'Не указана',
            'Experience': job['experience']['name'] if job['experience'] else 'Не указано',
            'Schedule': job['schedule']['name'] if job['schedule'] else 'Не указано',
            'Published_at': job['published_at'],
            'URL': job['alternate_url'],
            'Description': job.get('snippet', {}).get('requirement', 'Не указано'),
            'Key_skills': key_skills_str
        }
        job_data.append(job_info)

    return pd.DataFrame(job_data)


jobs_developer = get_multiple_pages_of_data("Разработчик", 2000)
df_developer = convert_to_dataframe(jobs_developer)
display(df_developer)
df_developer.to_csv("vacancies_developer.csv", index=False)
```

```python
# Настройка логирования
logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')

# Настройка ChromeOptions
options = webdriver.ChromeOptions()
options.add_argument('--headless')  # Режим без графического интерфейса
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
options.add_argument('--disable-gpu')
options.add_argument('--remote-debugging-port=9222')
options.add_argument('--disable-software-rasterizer')
options.add_argument(
    "user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 "
    "(KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
)
# Укажите путь к бинарному файлу Chrome, если требуется
options.binary_location = "/usr/bin/google-chrome"

# Инициализация драйвера
driver = webdriver.Chrome(
    service=Service(ChromeDriverManager().install()),
    options=options
)

def scroll_page():
    """Прокручивает страницу до самого низа, пока не загрузятся все вакансии."""
    last_height = driver.execute_script("return document.body.scrollHeight")
    while True:
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(2)  # Ожидание, чтобы страница успела подгрузить новые элементы
        new_height = driver.execute_script("return document.body.scrollHeight")
        if new_height == last_height:
            break
        last_height = new_height

def extract_detail_data(vacancy_url):
    """
    Открывает детальную страницу вакансии в новой вкладке и извлекает
    поля Experience, Schedule, Published_at, Key_skills и сам URL.
    """
    original_window = driver.current_window_handle
    # Значения по умолчанию
    experience, schedule, published_at, key_skills = None, None, None, None
    detail_url = vacancy_url

    try:
        driver.execute_script("window.open('');")
        new_window = [w for w in driver.window_handles if w != original_window][0]
        driver.switch_to.window(new_window)
        driver.get(vacancy_url)
        time.sleep(3)  # Ожидаем загрузку детальной страницы

        try:
          requirements_text = driver.find_element(By.CLASS_NAME, 'vacancy-requirements').text
          requirements_parts = requirements_text.split(", ")  # Разделяем строку по запятым

          schedule = requirements_parts[1] if len(requirements_parts) > 1 else None
          experience = requirements_parts[2] if len(requirements_parts) > 2 else None

        except Exception as e:
          print(f"Ошибка при извлечении требований: {e}")

        try:
            date_element = driver.find_element(By.CSS_SELECTOR, 'meta[itemprop="datePosted"]')
            published_at = date_element.get_attribute('content').split("T")[0]  # Берем только дату без времени
        except Exception as e:
            published_at = None
            print(f"Ошибка при извлечении даты: {e}")

        try:
            key_skills_elements = driver.find_elements(By.CLASS_NAME, 'vacancy-detail__key-skills')
            if key_skills_elements:
                key_skills = ", ".join(skill.text for skill in key_skills_elements)
            else:
                key_skills = "Не указано"
        except:
            key_skills = "Не указано"

        detail_url = vacancy_url

    except Exception as e:
        logging.error(f"Ошибка при извлечении детальных данных: {e}")
    finally:
        # Закрываем вкладку и возвращаемся в исходную
        driver.close()
        driver.switch_to.window(original_window)

    return {
        'Experience': experience,
        'Schedule': schedule,
        'Published_at': published_at,
        'URL': detail_url,
        'Key_skills': key_skills
    }

def extract_vacancy_data(vacancy_element):
    """
    Извлекает базовую информацию о вакансии (название, зарплату, компанию, описание),
    а затем вызывает extract_detail_data для дополнительных полей.
    """
    title = None
    salary_from, salary_to, salary_currency = None, None, None
    employer = None
    description = None
    vacancy_url = None
    # Извлекаем основные поля с краткой карточки
    try:
        title = vacancy_element.find_element(By.CLASS_NAME, 'vacancy-preview-card__title').text
    except:
        pass

    try:
        # Изначально обнуляем переменные
        salary_from = float('nan')
        salary_to = float('nan')
        salary_currency = None

        # Получаем текстовое значение, убираем неразрывные пробелы и обрезаем пробелы
        salary_raw = vacancy_element.find_element(By.CLASS_NAME, 'vacancy-preview-card__salary').text
        salary_text = salary_raw.replace('\u202f', ' ').strip().lower()

        # 1) "по договоренности"
        if "по договоренности" in salary_text:
            salary_currency = "По договоренности"

        # 2) "от x руб" (проверяем, что в начале есть "от ")
        elif salary_text.startswith("от "):
            # Убираем "от " из начала, получаем что-то вроде "100 000 руб" или "100000 руб."
            remainder = salary_text[3:].strip('.').strip()
            # Предположим, последние слова — это валюта ("руб" или "руб.")
            parts = remainder.split()
            # Если есть хотя бы одно число
            if len(parts) > 0:
                # Берём все части, кроме последней, как число
                # (например ["100", "000"] => "100000")
                if parts[-1] in ["руб", "руб."]:
                    # Последний элемент — валюта
                    salary_currency = "руб"
                    num_parts = parts[:-1]
                else:
                    # Если нет отдельной валюты, то всё — число
                    num_parts = parts
                    salary_currency = "руб"

                # Склеиваем число и переводим в int при желании
                if len(num_parts) > 0:
                    salary_from_str = ''.join(num_parts)
                    salary_from = float(salary_from_str) if salary_from_str.isdigit() else float('nan')

        # 3) "x - y руб" (диапазон)
        elif '—' in salary_text:
            # Например: "100 000 - 135 000 руб"
            left_part, right_part = salary_text.split('—', 1)
            left_part = left_part.strip()    # "100 000"
            right_part = right_part.strip()  # "135 000 руб"

            # Обрабатываем правую часть, чтобы отделить число от валюты
            # Разобьём по пробелам: ["135", "000", "руб"] или ["135000", "руб"]
            right_parts = right_part.split()
            num_parts = []
            for rp in right_parts:
                if rp in ["руб", "руб."]:
                    salary_currency = "руб"
                else:
                    num_parts.append(rp)

            # Склеиваем все части числа в правой части
            if len(num_parts) > 0:
                salary_to_str = ''.join(num_parts)
                salary_to = float(salary_to_str) if salary_to_str.isdigit() else float('nan')

            # Аналогично обрабатываем левую часть
            left_part = left_part.replace(' ', '')
            if left_part.isdigit():
                salary_from = float(left_part)
            else:
                salary_from = float('nan')

            # Если не нашли валюту в правой части — по умолчанию "руб"
            if not salary_currency:
                salary_currency = "руб"

        else:
            # Если нет "по договоренности", "от " и "-", то это может быть единичное значение, например "100 000 руб"
            parts = salary_text.split()
            if len(parts) > 0:
                # Предположим, что в конце может быть "руб" или "руб."
                if parts[-1] in ["руб", "руб."]:
                    salary_currency = "руб"
                    num_parts = parts[:-1]  # всё, кроме "руб..."
                else:
                    # Если не обнаружили явную валюту
                    num_parts = parts

                # Склеим всё в число
                if len(num_parts) > 0:
                    number_str = ''.join(num_parts).replace('.', '')
                    if number_str.isdigit():
                        salary_from = float(number_str)
                    else:
                        salary_from = float('nan')
                else:
                    salary_from = float('nan')

                # Для одного значения salary_to не задан -> NAN

        logging.info(f"Парсинг зарплаты: от {salary_from} до {salary_to} {salary_currency}")

    except Exception as e:
        logging.warning(f"Не удалось распарсить зарплату: {e}")
        pass

    try:
        employer = vacancy_element.find_element(By.CLASS_NAME, 'vacancy-preview-card__company-name').text
    except:
        pass

    try:
        description = vacancy_element.find_element(By.CLASS_NAME, 'vacancy-preview-card__description').text
    except:
        pass

    try:
      vacancy_url = vacancy_element.find_element(By.CLASS_NAME, 'vacancy-preview-card__title').find_element(By.TAG_NAME, 'a').get_attribute('href')
    except Exception as e:
      print(f"Ошибка при получении ссылки: {e}")

    # Извлекаем данные с детальной страницы (опыт, график, ключевые навыки и т.д.)
    if vacancy_url:
        detail_data = extract_detail_data(vacancy_url)
    else:
        detail_data = {
            'Experience': None,
            'Schedule': None,
            'Published_at': None
        }

    return {
        'Vacancy': title,
        'Salary_from': salary_from,
        'Salary_to': salary_to,
        'Salary_currency': salary_currency,
        'Employer_name': employer,
        'Experience': detail_data['Experience'],
        'Schedule': detail_data['Schedule'],
        'Published_at': detail_data['Published_at'],
        'URL': detail_data['URL'],
        'Description': description,
        'Key_skills': detail_data['Key_skills']
    }

def scrape_rabota_ru():
    """
    Переходит по страницам 1..N, формируя URL с номером страницы.
    На каждой странице прокручивает вниз, собирает вакансии и добавляет их в общий список.
    Останавливается, если вакансии на странице отсутствуют (или если достигнут лимит страниц).
    """
    data = []

    for page_number in range(1, 121):
        # Внимание: проверьте, верен ли URL с параметром ?page=.
        # Если нужно, поправьте на /vacancy?page= или другую структуру.
        url = f"https://www.rabota.ru/vacancy/?query=%D0%9C%D0%B5%D0%BD%D0%B5%D0%B4%D0%B6%D0%B5%D1%80&all_regions=1&page={page_number}"
        logging.info(f"Открываем страницу: {url}")
        driver.get(url)
        time.sleep(5)

        scroll_page()  # Прокрутка вниз для подгрузки всех вакансий

        vacancies = driver.find_elements(By.CLASS_NAME, 'vacancy-preview-card')
        if not vacancies:
            logging.info("Вакансий не найдено на странице. Останавливаемся.")
            break

        logging.info(f"Найдено {len(vacancies)} вакансий на странице {page_number}.")

        for vacancy in vacancies:
            vacancy_data = extract_vacancy_data(vacancy)
            data.append(vacancy_data)
            logging.info(f"Обработана вакансия: {vacancy_data['Vacancy']}")

    return data

# Запуск скрипта
try:
    scraped_data = scrape_rabota_ru()
    df = pd.DataFrame(scraped_data)
    df.to_csv('managers_par.csv', index=False)
    logging.info("Данные успешно сохранены в 'managers_par.csv'")
except Exception as e:
    logging.error(f"Произошла ошибка: {e}")
finally:
    driver.quit()
```

```python
# Настройка логирования
logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')

# Настройка ChromeOptions
options = webdriver.ChromeOptions()
options.add_argument('--headless')  # Режим без графического интерфейса
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
options.add_argument('--disable-gpu')
options.add_argument('--remote-debugging-port=9222')
options.add_argument('--disable-software-rasterizer')
options.add_argument(
    "user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 "
    "(KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
)
# Укажите путь к бинарному файлу Chrome, если требуется
options.binary_location = "/usr/bin/google-chrome"

# Инициализация драйвера
driver = webdriver.Chrome(
    service=Service(ChromeDriverManager().install()),
    options=options
)

def scroll_page():
    """Прокручивает страницу до самого низа, пока не загрузятся все вакансии."""
    last_height = driver.execute_script("return document.body.scrollHeight")
    while True:
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(2)  # Ожидание, чтобы страница успела подгрузить новые элементы
        new_height = driver.execute_script("return document.body.scrollHeight")
        if new_height == last_height:
            break
        last_height = new_height

def extract_detail_data(vacancy_url):
    """
    Открывает детальную страницу вакансии в новой вкладке и извлекает
    поля Experience, Schedule, Published_at, Key_skills и сам URL.
    """
    original_window = driver.current_window_handle
    # Значения по умолчанию
    experience, schedule, published_at, key_skills = None, None, None, None
    detail_url = vacancy_url

    try:
        driver.execute_script("window.open('');")
        new_window = [w for w in driver.window_handles if w != original_window][0]
        driver.switch_to.window(new_window)
        driver.get(vacancy_url)
        time.sleep(3)  # Ожидаем загрузку детальной страницы

        try:
          requirements_text = driver.find_element(By.CLASS_NAME, 'vacancy-requirements').text
          requirements_parts = requirements_text.split(", ")  # Разделяем строку по запятым

          schedule = requirements_parts[1] if len(requirements_parts) > 1 else None
          experience = requirements_parts[2] if len(requirements_parts) > 2 else None

        except Exception as e:
          print(f"Ошибка при извлечении требований: {e}")

        try:
            date_element = driver.find_element(By.CSS_SELECTOR, 'meta[itemprop="datePosted"]')
            published_at = date_element.get_attribute('content').split("T")[0]  # Берем только дату без времени
        except Exception as e:
            published_at = None
            print(f"Ошибка при извлечении даты: {e}")

        try:
            key_skills_elements = driver.find_elements(By.CLASS_NAME, 'vacancy-detail__key-skills')
            if key_skills_elements:
                key_skills = ", ".join(skill.text for skill in key_skills_elements)
            else:
                key_skills = "Не указано"
        except:
            key_skills = "Не указано"

        detail_url = vacancy_url

    except Exception as e:
        logging.error(f"Ошибка при извлечении детальных данных: {e}")
    finally:
        # Закрываем вкладку и возвращаемся в исходную
        driver.close()
        driver.switch_to.window(original_window)

    return {
        'Experience': experience,
        'Schedule': schedule,
        'Published_at': published_at,
        'URL': detail_url,
        'Key_skills': key_skills
    }

def extract_vacancy_data(vacancy_element):
    """
    Извлекает базовую информацию о вакансии (название, зарплату, компанию, описание),
    а затем вызывает extract_detail_data для дополнительных полей.
    """
    title = None
    salary_from, salary_to, salary_currency = None, None, None
    employer = None
    description = None
    vacancy_url = None
    # Извлекаем основные поля с краткой карточки
    try:
        title = vacancy_element.find_element(By.CLASS_NAME, 'vacancy-preview-card__title').text
    except:
        pass

    try:
        # Изначально обнуляем переменные
        salary_from = float('nan')
        salary_to = float('nan')
        salary_currency = None

        # Получаем текстовое значение, убираем неразрывные пробелы и обрезаем пробелы
        salary_raw = vacancy_element.find_element(By.CLASS_NAME, 'vacancy-preview-card__salary').text
        salary_text = salary_raw.replace('\u202f', ' ').strip().lower()

        # 1) "по договоренности"
        if "по договоренности" in salary_text:
            salary_currency = "По договоренности"

        # 2) "от x руб" (проверяем, что в начале есть "от ")
        elif salary_text.startswith("от "):
            # Убираем "от " из начала, получаем что-то вроде "100 000 руб" или "100000 руб."
            remainder = salary_text[3:].strip('.').strip()
            # Предположим, последние слова — это валюта ("руб" или "руб.")
            parts = remainder.split()
            # Если есть хотя бы одно число
            if len(parts) > 0:
                # Берём все части, кроме последней, как число
                # (например ["100", "000"] => "100000")
                if parts[-1] in ["руб", "руб."]:
                    # Последний элемент — валюта
                    salary_currency = "руб"
                    num_parts = parts[:-1]
                else:
                    # Если нет отдельной валюты, то всё — число
                    num_parts = parts
                    salary_currency = "руб"

                # Склеиваем число и переводим в int при желании
                if len(num_parts) > 0:
                    salary_from_str = ''.join(num_parts)
                    salary_from = float(salary_from_str) if salary_from_str.isdigit() else float('nan')

        # 3) "x - y руб" (диапазон)
        elif '—' in salary_text:
            # Например: "100 000 - 135 000 руб"
            left_part, right_part = salary_text.split('—', 1)
            left_part = left_part.strip()    # "100 000"
            right_part = right_part.strip()  # "135 000 руб"

            # Обрабатываем правую часть, чтобы отделить число от валюты
            # Разобьём по пробелам: ["135", "000", "руб"] или ["135000", "руб"]
            right_parts = right_part.split()
            num_parts = []
            for rp in right_parts:
                if rp in ["руб", "руб."]:
                    salary_currency = "руб"
                else:
                    num_parts.append(rp)

            # Склеиваем все части числа в правой части
            if len(num_parts) > 0:
                salary_to_str = ''.join(num_parts)
                salary_to = float(salary_to_str) if salary_to_str.isdigit() else float('nan')

            # Аналогично обрабатываем левую часть
            left_part = left_part.replace(' ', '')
            if left_part.isdigit():
                salary_from = float(left_part)
            else:
                salary_from = float('nan')

            # Если не нашли валюту в правой части — по умолчанию "руб"
            if not salary_currency:
                salary_currency = "руб"

        else:
            # Если нет "по договоренности", "от " и "-", то это может быть единичное значение, например "100 000 руб"
            parts = salary_text.split()
            if len(parts) > 0:
                # Предположим, что в конце может быть "руб" или "руб."
                if parts[-1] in ["руб", "руб."]:
                    salary_currency = "руб"
                    num_parts = parts[:-1]  # всё, кроме "руб..."
                else:
                    # Если не обнаружили явную валюту
                    num_parts = parts

                # Склеим всё в число
                if len(num_parts) > 0:
                    number_str = ''.join(num_parts).replace('.', '')
                    if number_str.isdigit():
                        salary_from = float(number_str)
                    else:
                        salary_from = float('nan')
                else:
                    salary_from = float('nan')

                # Для одного значения salary_to не задан -> NAN

        logging.info(f"Парсинг зарплаты: от {salary_from} до {salary_to} {salary_currency}")

    except Exception as e:
        logging.warning(f"Не удалось распарсить зарплату: {e}")
        pass

    try:
        employer = vacancy_element.find_element(By.CLASS_NAME, 'vacancy-preview-card__company-name').text
    except:
        pass

    try:
        description = vacancy_element.find_element(By.CLASS_NAME, 'vacancy-preview-card__description').text
    except:
        pass

    try:
      vacancy_url = vacancy_element.find_element(By.CLASS_NAME, 'vacancy-preview-card__title').find_element(By.TAG_NAME, 'a').get_attribute('href')
    except Exception as e:
      print(f"Ошибка при получении ссылки: {e}")

    # Извлекаем данные с детальной страницы (опыт, график, ключевые навыки и т.д.)
    if vacancy_url:
        detail_data = extract_detail_data(vacancy_url)
    else:
        detail_data = {
            'Experience': None,
            'Schedule': None,
            'Published_at': None
        }

    return {
        'Vacancy': title,
        'Salary_from': salary_from,
        'Salary_to': salary_to,
        'Salary_currency': salary_currency,
        'Employer_name': employer,
        'Experience': detail_data['Experience'],
        'Schedule': detail_data['Schedule'],
        'Published_at': detail_data['Published_at'],
        'URL': detail_data['URL'],
        'Description': description,
        'Key_skills': detail_data['Key_skills']
    }

def scrape_rabota_ru():
    """
    Переходит по страницам 1..N, формируя URL с номером страницы.
    На каждой странице прокручивает вниз, собирает вакансии и добавляет их в общий список.
    Останавливается, если вакансии на странице отсутствуют (или если достигнут лимит страниц).
    """
    data = []
    # Можно задать нужное кол-во страниц вручную или сделать бесконечный цикл while True.
    # Здесь пример с диапазоном 1..50.
    for page_number in range(1, 51):
        # Внимание: проверьте, верен ли URL с параметром ?page=.
        # Если нужно, поправьте на /vacancy?page= или другую структуру.
        url = f"https://www.rabota.ru/vacancy/?query=Hr-%D0%BC%D0%B5%D0%BD%D0%B5%D0%B4%D0%B6%D0%B5%D1%80&all_regions=1&page={page_number}"
        logging.info(f"Открываем страницу: {url}")
        driver.get(url)
        time.sleep(5)

        scroll_page()  # Прокрутка вниз для подгрузки всех вакансий

        vacancies = driver.find_elements(By.CLASS_NAME, 'vacancy-preview-card')
        if not vacancies:
            logging.info("Вакансий не найдено на странице. Останавливаемся.")
            break

        logging.info(f"Найдено {len(vacancies)} вакансий на странице {page_number}.")

        for vacancy in vacancies:
            vacancy_data = extract_vacancy_data(vacancy)
            data.append(vacancy_data)
            logging.info(f"Обработана вакансия: {vacancy_data['Vacancy']}")

    return data

# Запуск скрипта
try:
    scraped_data = scrape_rabota_ru()
    df = pd.DataFrame(scraped_data)
    df.to_csv('hr_par.csv', index=False)
    logging.info("Данные успешно сохранены в 'hr_par.csv'")
except Exception as e:
    logging.error(f"Произошла ошибка: {e}")
finally:
    driver.quit()
```
